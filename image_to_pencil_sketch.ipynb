import python **libaries*

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Get the image path from the user
image_path = "/content/image.jpg"

# Read the image
image = cv2.imread(image_path)

# Display the original RGB image
cv2_imshow(image)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray_image)

# Invert the grayscale image
inverted_gray_image = 255 - gray_image

# Blur the inverted grayscale image
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)

# Display the blurred image
cv2_imshow(blurred_image)

# Invert the blurred image
inverted_blurred_image = 255 - blurred_image

# Blend the blurred and inverted images
sketch = cv2.addWeighted(gray_image, 0.7, inverted_blurred_image, 0.3, 0)

# Save the sketch
cv2.imwrite("/content/sketch.jpg", sketch)

# Display the sketch
cv2_imshow(sketch)

print("The pencil sketch has been saved as 'sketch.jpg'.")
```
